```python
def model_development(df, target_column, sampling_technique, model_name):
    print(f'DATA SPLITTING & SAMPLING\n')
    print('started---')
    X_train_resampled, X_test, y_train_resampled, y_test = split_and_sample_data(df, target_column, sampling_technique, random_state=42)
    X_train =  X_train_resampled
    y_train = y_train_resampled
    print('completed!')
    if model_name == 'rfc':
        model = RandomForestClassifier(random_state=42)
    elif model_name == 'gbc':
        model = GradientBoostingClassifier(random_state=42)
    elif model_name == 'mlp':
        model = MLPClassifier(max_iter=1000)
    
    print(f'\nDATA PIPELINE CREATION\n')
    print('started---')
    pipeline = create_pipeline(model)
    print('completed!')
    print(f'\nCROSS VALIDATION\n')
    print('started---')
    Matrix_df = cross_validation(pipeline, X_train, y_train, n_splits=10)
    print(Matrix_df)
    print('started---')
    print(f'\nMODEL TRAINING ON FULL DATASET\n')
    print('started---')
    pipeline.fit(X_train, y_train) 
    print('completed!')
   
    print(f'\nMODEL EVALUATION\n')
    print('started---')
    classification_report_output, roc_auc_output = evaluate_model(pipeline, X_test, y_test)
    plot_classification_report(pipeline, X_train, y_train, X_test, y_test, title="Classification Report")
    plot_confusion_matrix(pipeline, X_test, y_test)
    plot_class_prediction_error(pipeline, X_train, y_train, X_test, y_test)
    plot_roc_auc(pipeline, X_train, y_train, X_test, y_test)
    plot_precision_recall_vs_threshold(pipeline, X_test, y_test)
    print('completed!')
    
    joblib.dump(pipeline, f'{model_name}_{sampling_technique}_pipeline.pkl')
    print('trained model saved!')
    print(f'model name:{model_name}_{sampling_technique}_pipeline.pkl')
    
    # Return the trained pipeline
    return pipeline
```

```python
df = df
target_column = 'isFraud'
sampling_technique = 'adasyn'
model_name = 'rfc'
model_development(df, target_column, sampling_technique, model_name)
```
